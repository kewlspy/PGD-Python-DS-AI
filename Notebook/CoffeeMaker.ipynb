```python
import pandas as pd

# Define the data
coffee_data = {'Id':range(1,11),
    'Coffee Name': ['Espresso', 'Latte', 'Cappuccino', 'Americano', 'Mocha', 'Macchiato', 'Flat White', 'Affogato', 'Cortado', 'Irish Coffee'],
    'Price (USD)': [2.5, 3.5, 3.0, 2.0, 4.0, 3.0, 3.5, 5.0, 3.0, 6.0],
    'Size': ['Small', 'Medium', 'Medium', 'Large', 'Medium', 'Small', 'Medium', 'Small', 'Small', 'Large'],
    'Caffeine Content (mg)': [75, 80, 85, 100, 90, 75, 80, 90, 85, 95],
    'Type': ['Espresso', 'Latte', 'Cappuccino', 'Americano', 'Mocha', 'Macchiato', 'Flat White', 'Affogato', 'Cortado', 'Specialty']
}

# Create the DataFrame
coffee_df = pd.DataFrame(coffee_data)

# Display the DataFrame
coffee_df
```

```python
def start(): 
    orderList=[]
    orderDetail=[]
    name=input("\nHello! Press '1' to see Menu and '0' to Exit.")
    if name=="1":
        
        def addOrder():
            print("\n")
            print(coffee_df)
            order=int(input("\nEnter your desired coffee number."))
            if order in coffee_df['Id'].values:
                orderList.append(order)
                print("\nYour Order List\n")
                for coffee_id in orderList:
                    selected_coffee = coffee_df[coffee_df['Id'] == coffee_id]
                    print(selected_coffee)
                
                addorder=int(input("\nDo you want to order more? If Yes enter '1' If No then enter '0' : "))
                if addorder==1:
                    addOrder()
                else:
                    total=0
                    for coffee_id in orderList:
                        total += coffee_df.loc[coffee_df['Id'] == coffee_id, 'Price (USD)'].values[0]
                    print(f"\nThe total cost of your order is: ${total:.2f}")
                    print("\nThank you for having coffee with us!")
        addOrder()
   

    else :
        start()


start()
```
